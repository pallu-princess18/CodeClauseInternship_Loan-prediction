```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import numpy as np
import pandas as pd
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('train.csv')
df.head(10)
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
nunique = df_train.nunique()
df = df_train[[col for col in df_train if nunique[col] > 1 and nunique[col] <5]] 
df
```

```python
df_train.isna().sum()
```

```python
for column in df.columns:
    df_train[column].fillna( method ='ffill', inplace = True)
```

```python
df_train['Loan_Amount_Term'] = df_train['Loan_Amount_Term'].replace(np.nan,df_train['Loan_Amount_Term'].mean())
```

```python
df_train['LoanAmount'] = df_train['LoanAmount'].replace(np.nan,df_train['LoanAmount'].mean())
```

```python
df_train['Loan_Status'] = df_train['Loan_Status'].map({'Y': 1, 'N': 0})
```

```python
df_train['Loan_Amount_Term'].unique()
```

```python
# Loan_ID
X = df_train.drop('Loan_ID', axis=1)
```

```python
X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']
```

```python
X.fillna( method ='ffill', inplace = True)
```

```python
X = pd.get_dummies(X , drop_first = True)
X 
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X, y , test_size = 0.25 , random_state = 42, stratify=y)
```

```python
y_test.value_counts()
```

```python
y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train_scaled,y_train)
y_pred = lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train_scaled,y_train)
y_pred = rf.predict(x_test_scaled)
print('evaluation on training set: {}'.format(rf.score(x_train_scaled,y_train)))
print('evaluation on testing set: {}'.format(rf.score(x_test_scaled,y_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train_scaled,y_train)
y_pred = knn.predict(x_test_scaled)
print('evaluation on training set: {}'.format(knn.score(x_train_scaled,y_train)))
print('evaluation on testing set: {}'.format(knn.score(x_test_scaled,y_test)))
```
